```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
import chromedriver_autoinstaller
import pandas as pd
import csv
import time
```

```python
chromedriver_autoinstaller.install()

browser = webdriver.Chrome()

data = []

for index in range (1,2):
    browser.get(f'https://www.tiket.com/sewa-mobil/cari?regional_id=1ad81f07-0db8-4ae6-99af-6576d6b0f45f&start_date=2023-06-14&end_date=2023-06-14&qty=1&set_driver=withDriver')
    time.sleep(5)
    soup = BeautifulSoup(browser.page_source,"lxml")
    rows = soup.find_all('div', class_='vendors-info-box search-car-list')
```

```python
for row in rows:
        Nama_Travel = row.find(class_='vendor-name')
        if Nama_Travel:
            Nama_Travel = Nama_Travel.text
            
        Kapasitas_Penumpang = row.find(class_='passenger-span')
        if Kapasitas_Penumpang:
            Kapasitas_Penumpang = Kapasitas_Penumpang.text
        
        Kapasitas_Bagasi = row.find(class_='baggage-span')
        if Kapasitas_Bagasi:
            Kapasitas_Bagasi = Kapasitas_Bagasi.text
        
        df = {
            'Nama Travel': Nama_Travel,
            'Kapasitas Penumpang': Kapasitas_Penumpang,
            'Kapasitas Bagasi': Kapasitas_Bagasi
            }
        data.append(df)
```

```python
df = pd.DataFrame(data)
```
